```python
import os
root_path = os.getcwd()
image_path = root_path+"/test_imgs/test_1.png"
```

```python
#Read image with open cv and default color channel is BGR , not RGB
import cv2

bgr_img = cv2.imread(image_path)
print(type(bgr_img))
if(bgr_img is not None):
    width,height,channel = bgr_img.shape
    print(width,height)
    print(channel)
```

```python
# Covert BGR to RGB and show image with Matplot 
import matplotlib.pyplot as plt

rgb_img = cv2.cvtColor(bgr_img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img[:,:,0])
```

```python
# RGB to Gray Image
gray_img = cv2.cvtColor(rgb_img,cv2.COLOR_RGB2GRAY)
print(gray_img.shape)
plt.imshow(gray_img)
```

```python
#Blur image
gray_blurred = cv2.blur(gray_img, (5, 5))
plt.imshow(gray_blurred)
```

```python
# Circle Detection
detected_circles = cv2.HoughCircles(gray_img,  
                   cv2.HOUGH_GRADIENT,dp = 1, minDist  = 2, param1 = 50, 
               param2 = 50, minRadius = 1, maxRadius = 40) 
print(detected_circles)
```

```python
# Draw Circle at a place where cirecle is detected
import numpy as np
if detected_circles is not None:
    detected_circles = np.uint16(np.round(detected_circles))
    print(detected_circles)
    for pt in detected_circles[0,:]: # [[[]]] to [[]]
        a,b,r = pt[0],pt[1],pt[2]
        cv2.circle(rgb_img,(a,b),radius=r,color=(0,255,0),thickness=1)
    
plt.imshow(rgb_img)

        
```
